### Contours
A contour is a curve that traces the boundary of a shape or region in an image.  
Pass the edges found by `Canny` into OpenCV's `findContours` function.

```python
img = cv2.imread('aiclass.png')
edges = cv2.Canny(img, 100, 200)
```

```python
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

What's in contours?  Check one out.

 Each item in `contours` is a list of points (x,y coords) on a contour.

```python
out_img=cv2.drawContours(img, contours, -1, (0,255,0), 2)
cv2.imshow("Contours", out_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Haar cascades for face detection
We will just take a quick look at this, treating it as a black-box. OpenCV2 comes with pre-trained classifier models. These are stored in xml files normally in `\\site-packages\\cv2\\data\\`. (To find their location, in Python after importing cv2, type `print(cv2.data.haarcascades)`.)  We will use the *frontalface* classifier to detect faces. 

```python
# YOUR CODE HERE
```

```python
# Load pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load an image
img = cv2.imread('aiclass.png')  # or use a.png, any image with a face
```

Haar classifiers are trained on grayscale images, focusing on differences in brightness over pixels rather than color information. So like the edge detection, let's convert the image to grayscale for face detection.

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

Now run the face detection model on the grayscale image.

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
```

What did it return?  What are faces?  What data type is it / what are its contents?

```python
# YOUR CODE HERE
```

Looks like it is a numpy array of 4-element vectors. These are rectangles:  (x, y, w, h) - top left x and y, width and height of rectangle.

Finally, draw rectangles on the face locations and display them over the original image. We will use `cv2.rectangle` (type `cv2.rectangle?` to see its documentation).

```python
# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Show image
cv2.imshow('Faces', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
