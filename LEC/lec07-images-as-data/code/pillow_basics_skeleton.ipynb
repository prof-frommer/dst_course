Tile the small image across a larger canvas.

```python
wallpaper = Image.new('RGB', img.size)
w, h = small.size
for y in range(0, img.size[1], h):
    for x in range(0, img.size[0], w):
        wallpaper.paste(small, (x, y))
wallpaper
```
