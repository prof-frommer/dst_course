## Creating sliders
You can create your own custom sliders without using `interact`.

```python
slider = widgets.FloatSlider(
    value=7.5,
    min=5.0,
    max=10.0,
    step=0.1,
    description='Input:',
)
slider
# note arrow keys work on it
```

You can capture values from your slider by defining a function that will **observe** it.  
First note the value of the slider is stored in `slider.value`.

```python
square = slider.value * slider.value

def handle_change(change):
    global square
    square = change.new * change.new
    
slider.observe(handle_change, 'value')
```

```python
square
```

We have to re-run the square cell to display the updated value. We will see an alternative later where that is not necessary.

## Simple plotter

```python
import matplotlib.pyplot as plt
import numpy as np

def f(m, b):
    x = np.linspace(-10, 10, num=1000)
    
    plt.figure()
    plt.plot(x, m * x + b)
    plt.grid()
    plt.ylim(-5, 5)
    plt.show()

    
    plt.figure()


interact(f, m=(-2.0, 2.0), b=(-3, 3, 0.5));
```

```python
# optional - make sliders more explicit

interact(
    f,
    m=widgets.FloatSlider(min=-2, max=2, step=0.1, value=1),
    b=widgets.FloatSlider(min=-3, max=3, step=0.5, value=0),
)
```

### Another example - sin curve with controllable parameters
Note the use of  `fixed` to not have a function input be made a slider by `interact`.

```python
x = np.linspace(0, 4 * np.pi, 100)
k = 1
p = np.pi

def f(x, k, p):
    def inner_f(x, k, p):
        return np.sin(k*x -p)
    fig, ax = plt.subplots()
    ax.grid()
    ax.plot(x, inner_f(x, k, p))
interact(f, x=fixed(x),k=(.5,2), p=(0,np.pi))
```

### Widgets Working with Dataframes

Load some libraries and a dataframe of Calculus grades.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# sample data
df = pd.read_csv('calc1.csv')
df = df.drop('first name',axis=1)
df
```

Now plot a histogram of a chosen column.

```python
# widgets
column = widgets.Dropdown(options=df.columns, description="Column:")
min_val = widgets.FloatSlider(min=0, max=1, value=0, step=.1,description="Min:")
max_val = widgets.FloatSlider(min=0, max=1, value=1, step=.1,description="Max:")

output = widgets.Output()

def update(change=None):
    output.clear_output()
    
    col = column.value
    filtered = df[(df[col] >= min_val.value) & (df[col] <= max_val.value)]
    
    with output:
        plt.figure()
        plt.hist(filtered[col], bins=20)
        plt.title(f"{col} (filtered)")
        plt.show()

# trigger updates
column.observe(update, names='value')
min_val.observe(update, names='value')
max_val.observe(update, names='value')

# initial draw
update()

display(column, min_val, max_val, output)
```

### view the data in table form as well

```python
# widgets
columns = widgets.SelectMultiple(
    options=df.columns,
    value=[df.columns[0]],
    description="Columns:"
)

plot_col = widgets.Dropdown(
    options=df.columns,
    description="Plot:"
)

min_val = widgets.FloatSlider(min=0, max=1, value=0, step=.1, description="Min:")
max_val = widgets.FloatSlider(min=0, max=1, value=1, step=.1,description="Max:")

row_limit = widgets.IntSlider(min=5, max=50, value=10, description="Rows:")

output = widgets.Output()

def update(change=None):
    output.clear_output()
    
    # filtering based on selected plot column
    col = plot_col.value
    filtered = df[(df[col] >= min_val.value) & (df[col] <= max_val.value)]
    
    with output:
        # ---- TABLE ----
        display(filtered[list(columns.value)].head(row_limit.value))
        
        # ---- PLOT ----
        plt.figure()
        plt.hist(filtered[col], bins=20)
        plt.title(f"{col} (filtered)")
        plt.show()

# connect widgets
for w in [columns, plot_col, min_val, max_val, row_limit]:
    w.observe(update, names='value')

# initial draw
update()

# layout
controls = widgets.VBox([columns, plot_col, min_val, max_val, row_limit])
display(widgets.HBox([controls, output]))
```
